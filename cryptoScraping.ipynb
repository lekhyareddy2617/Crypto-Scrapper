```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
website='https://www.coingecko.com/en'
#req data: name, price, 1h change, 24h change, 7d change, 24h volume, market cap
```

```python
#get request
response=requests.get(website)
response.status_code #200 means we have made succesful http request
```

```python
soup=BeautifulSoup(response.content,'html.parser') #this soup object gives access to html objects of the page
#soup
results=soup.find('table', {'class':'table-scrollable'}).find('tbody').find_all('tr')
#on using inspect element we see that the page has tables with rows which is what we do to find results
len(results)
```

```python
#name
results[0].find('a', {'class':'tw-hidden lg:tw-flex font-bold tw-items-center tw-justify-between'}).get_text().strip() #for the first row
#the argument of find() is obtained using inspect element

#price
results[0].find('td', {'class':'td-price'}).get_text().strip() #for first row

#1h change
results[0].find('td', {'class':'td-change1h'}).get_text().strip()

#24h change
results[0].find('td', {'class':'td-change24h'}).get_text().strip()

#7d change
results[0].find('td', {'class':'td-change7d'}).get_text().strip()

#volume
results[0].find('td', {'class':'td-liquidity_score'}).get_text().strip()

#market cap
results[0].find('td', {'class':'td-market_cap'}).get_text().strip()
```

```python
#100 rows, 7 data points
name=[]
price=[]
change1h=[]
change24h=[]
change7d=[]
volume24h=[]
marketcap=[]

for result in results:
    #name
    try:
        name.append(result.find('a', {'class':'tw-hidden lg:tw-flex font-bold tw-items-center tw-justify-between'}).get_text().strip())
    except:
        name.append('NA')
    #price
    try:
        price.append(result.find('td', {'class':'td-price'}).get_text().strip())
    except:
        price.append('NA')
    #change1h
    try:
        change1h.append(result.find('td', {'class':'td-change1h'}).get_text().strip())
    except:
        change1h.append('NA')
    #change24h
    try:
        change24h.append(result.find('td', {'class':'td-change24h'}).get_text().strip())
    except:
        change24h.append('NA')
    #change7d
    try:
        change7d.append(result.find('td', {'class':'td-change7d'}).get_text().strip())
    except:
        change7d.append('NA')
    #volume24h
    try:
        volume24h.append(result.results[0].find('td', {'class':'td-liquidity_score'}).get_text().strip())
    except:
        volume24h.append('NA')
    #marketcap
    try:
        marketcap.append(result.find('td', {'class':'td-market_cap'}).get_text().strip())
    except:
        marketcap.append('NA')

len(name)
```

```python
crypto_dataframe=pd.DataFrame({'Coin Name':name, 'Price':price, 'Change1h':change1h,'Change24h':change24h,'Change7d':change7d,'Volume':volume24h,'MarketCap':marketcap})
crypto_dataframe
```

```python
#100 rows, 7 data points, 10 pages
name=[]
price=[]
change1h=[]
change24h=[]
change7d=[]
volume24h=[]
marketcap=[]

for page in range (1,11):
    print(page)
    website='https://www.coingecko.com/en?page='+str(page)
    response = requests.get(website)
    soup=BeautifulSoup(response.content,'html.parser')
    results=soup.find('table', {'class':'table-scrollable'}).find('tbody').find_all('tr')

    for result in results:
        #name
        try:
            name.append(result.find('a', {'class':'tw-hidden lg:tw-flex font-bold tw-items-center tw-justify-between'}).get_text().strip())
        except:
            name.append('NA')
        #price
        try:
            price.append(result.find('td', {'class':'td-price'}).get_text().strip())
        except:
            price.append('NA')
        #change1h
        try:
            change1h.append(result.find('td', {'class':'td-change1h'}).get_text().strip())
        except:
            change1h.append('NA')
        #change24h
        try:
            change24h.append(result.find('td', {'class':'td-change24h'}).get_text().strip())
        except:
            change24h.append('NA')
        #change7d
        try:
            change7d.append(result.find('td', {'class':'td-change7d'}).get_text().strip())
        except:
            change7d.append('NA')
        #volume24h
        try:
            volume24h.append(result.results[0].find('td', {'class':'td-liquidity_score'}).get_text().strip())
        except:
            volume24h.append('NA')
        #marketcap
        try:
            marketcap.append(result.find('td', {'class':'td-market_cap'}).get_text().strip())
        except:
            marketcap.append('NA')
```

```python
crypto_dataframe=pd.DataFrame({'Coin Name':name, 'Price':price, 'Change1h':change1h,'Change24h':change24h,'Change7d':change7d,'Volume':volume24h,'MarketCap':marketcap})
crypto_dataframe
```
